### Inflection is the change in the form of a word (in English, usually the addition of endings) to mark such distinctions as tense, person, number, gender, mood, voice, and case.

<bt></br>

### Lemmatization is the process of converting a word to its base form

<br></br>
<bt></br>




<br></br>
<br></br>

### Lemmatization is done in order to remove convert different form of a word into its base form, once done, different forms of a word are treated the same no matter the form they are being used in.

<br></br>
<br></br>

### Lemmatization is widely used in tagging systems, indexing, SEOs, Web search results, and information retrieval.


<br></br>
### Code Implementation

```python
#Importing spacy library 

import spacy
```

```python
# Loading the Rules

nlp = spacy.load('en_core_web_sm')
```

```python
text = """I moved from a 6s to the Xs, the speed and feel of the phone is brilliant.\
For everyone who thinks it’s overpriced, yes we pay for a brand and it’s a choice we make."""
```

```python
document = nlp(text)
```

```python
# Converting Tokens to their base form

for i in document:
    if i.is_stop != True and i.is_punct != True:
        print(i.text, '-->' ,i.lemma_)
```
