### Code Implementation

```python
#Importing spacy library 

import spacy
```

```python
# Loading the Rules

nlp = spacy.load('en_core_web_sm')
```

```python
text = """I moved from a 6s to the Xs, the speed and feel of the phone is brilliant.\
For everyone who thinks it’s overpriced, yes we pay for a brand and it’s a choice we make."""
```

```python
document = nlp(text)
```

```python
# Importing stopwords
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
#Printing all the stopwords present inside english small package
print(len(STOP_WORDS))
print(STOP_WORDS)
```

#### Checking whether any word is stop word or not

```python
# Checking if a word is stopword or not
nlp.vocab['the'].is_stop
```

```python
nlp.vocab['machine'].is_stop
```

Here, <b>the</b> is a stop word while <b>machine</b> is not a stop word

### Applying Stop Words over sample text

```python
# Iterating over tokenized text 
for i in document:
    if i.is_stop!=True and i.is_punct!=True:          #Printing if word is not a stopword and not a punctuation
        print(i)
```

#### Adding custom stopword to the default stopwords

```python
# Adding custon stopwords to the default stopwords
nlp.vocab['machine'].is_stop = True
```
