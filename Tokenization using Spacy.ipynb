# <u>Tokenization Using Spacy</u>

##  Tokenization is the step in Natural Language Processing where we break down large text into atomic units i.e. words or sentences.


## It is easy for feeding smaller chunks into machine for processing then passing text as whole. 



## Tokenization makes Natural Language Processing Fast and scalable. 

## Tokenization does this task by locating word boundaries. Ending point of a word and beginning of the next word is called word boundaries.


## Tokenization in NLP is of Two Types. They are Sentence Tokenization and Word Tokenization

. 

.

### Importing Spacy and English Rules and Regulations

```python
#Importing spacy library 

import spacy
```

```python
#Download English rules and regulations 


!python -m spacy download en_core_web_sm
```

```python
# Loading the Rules

nlp = spacy.load('en_core_web_sm')
```

```python
type(nlp)
```

```python
text = """I moved from a 6s to the Xs, the speed and feel of the phone is brilliant.\
For everyone who thinks its overpriced, yes we pay for a brand and its a choice we make."""
```

```python
# Converting text to tokenized document


document = nlp(text)
```

```python
type(document)
```

## Implementing Sentence Tokenization

```python
# Iterating over document.sents separated using punctuations. 

for i in document.sents:
    print(i)
```

## Implementing Word Tokenization

```python
# Iterating directly over tokenized document, containing objects like punctuations, alphabets, numbers etc.

for i in document:
    print(i, end=', ')
```
