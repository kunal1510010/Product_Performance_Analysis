# <u>TF-IDF</u>

### <u>Code Implementation</u>

```python
# Importing TF-IDF Vectorizer

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Defining a corpus

sentences = ["the dog is sitting in the porch",
            "the cat is sitting on my lap"]
```

```python
# Defining Vectorizer object

vectorizer = TfidfVectorizer(analyzer = 'word',stop_words='english')
```

```python
# Fit Transforming Corpus to vectorizer

response = vectorizer.fit_transform(sentences)
```

```python
# Getting feature names

feature_names = vectorizer.get_feature_names()
print(feature_names)
```

```python
print(response)
```

```python
# Transform function to convert text to array

vectorizer.transform(["the cat is sitting on my lap"]).toarray()
```

```python
# Dense function returning a sparse matrix
dense = response.todense()

dense
```

```python

#Then we convert that dense matrix into a list(2D list) using tolist() functionality.
denselist = dense.tolist()
print(denselist)
```

```python
# Pandas library for tabular format representation 

import pandas as pd
df = pd.DataFrame(denselist, columns = feature_names)
df
```
